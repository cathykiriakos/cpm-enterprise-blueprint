## 7. Business Impact: Prediction Results

```python
# Generate sample predictions showing business value
np.random.seed(42)

n_constituents = 1000
predictions = pd.DataFrame({
    'constituent_id': [f'C-{i:05d}' for i in range(n_constituents)],
    'churn_risk': np.random.beta(2, 5, n_constituents),  # More low-risk
    'upgrade_propensity': np.random.beta(3, 5, n_constituents),
    'lifetime_value': np.random.lognormal(mean=4, sigma=1, size=n_constituents)
})

# Categorize
predictions['churn_tier'] = pd.cut(predictions['churn_risk'], 
                                    bins=[0, 0.3, 0.6, 0.8, 1], 
                                    labels=['Low', 'Medium', 'High', 'Critical'])
predictions['upgrade_potential'] = pd.cut(predictions['upgrade_propensity'],
                                          bins=[0, 0.3, 0.6, 0.8, 1],
                                          labels=['Low', 'Medium', 'High', 'Critical'])

# Calculate segments
churn_counts = predictions['churn_tier'].value_counts().sort_index()
upgrade_counts = predictions['upgrade_potential'].value_counts().sort_index()

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Constituent Churn Risk Distribution',
        'Upgrade Opportunity Distribution',
        'Churn Risk vs Lifetime Value',
        'Business Outcome Summary'
    ),
    specs=[
        [{'type': 'pie'}, {'type': 'pie'}],
        [{'type': 'scatter'}, {'type': 'indicator'}]
    ],
    vertical_spacing=0.15,
    horizontal_spacing=0.15
)

# Churn risk pie
colors_churn = {'Low': '#4CAF50', 'Medium': '#FFC107', 'High': '#FF9800', 'Critical': '#F44336'}
fig.add_trace(
    go.Pie(
        labels=churn_counts.index,
        values=churn_counts.values,
        marker=dict(colors=[colors_churn[x] for x in churn_counts.index]),
        hovertemplate='<b>%{label}</b><br>Count: %{value}<br>Percentage: %{percent}<extra></extra>',
        textinfo='label+percent'
    ),
    row=1, col=1
)

# Upgrade potential pie
colors_upgrade = {'Low': '#90EE90', 'Medium': '#FFF59D', 'High': '#FFD54F', 'Critical': '#FFAB91'}
fig.add_trace(
    go.Pie(
        labels=upgrade_counts.index,
        values=upgrade_counts.values,
        marker=dict(colors=[colors_upgrade[x] for x in upgrade_counts.index]),
        hovertemplate='<b>%{label}</b><br>Count: %{value}<br>Percentage: %{percent}<extra></extra>',
        textinfo='label+percent'
    ),
    row=1, col=2
)

# Risk vs Value scatter
fig.add_trace(
    go.Scatter(
        x=predictions['churn_risk'],
        y=predictions['lifetime_value'],
        mode='markers',
        marker=dict(
            size=5,
            color=predictions['churn_risk'],
            colorscale='RdYlGn_r',
            showscale=True,
            colorbar=dict(title='Churn Risk')
        ),
        text=[f'ID: {id}<br>Risk: {risk:.2f}<br>Value: ${val:,.0f}' 
              for id, risk, val in zip(predictions['constituent_id'], 
                                       predictions['churn_risk'],
                                       predictions['lifetime_value'])],
        hovertemplate='%{text}<extra></extra>'
    ),
    row=2, col=1
)

# Key insight
high_risk_high_value = len(predictions[(predictions['churn_risk'] > 0.6) & (predictions['lifetime_value'] > 100)])
fig.add_trace(
    go.Indicator(
        mode='number+delta',
        value=high_risk_high_value,
        title={'text': 'High-Value At-Risk\nConstituents'},
        delta={'reference': 0, 'relative': False},
        number=dict(font=dict(size=48), suffix=' constituents')
    ),
    row=2, col=2
)

fig.update_xaxes(title_text='Churn Risk Score', row=2, col=1, range=[0, 1])
fig.update_yaxes(title_text='Lifetime Value ($)', row=2, col=1, type='log')

fig.update_layout(
    title_text='ðŸ’° Prediction Results: Segments & Opportunities',
    height=800,
    showlegend=False
)

fig.show()

print(f"\nðŸ“Š Key Findings from {n_constituents:,} Constituents:")
print(f"\nðŸ”´ CHURN RISK:")
for tier in ['Critical', 'High', 'Medium', 'Low']:
    count = (predictions['churn_tier'] == tier).sum()
    pct = 100 * count / n_constituents
    print(f"   {tier:12}: {count:4} constituents ({pct:5.1f}%) - PRIORITY: {'ðŸ”´ URGENT' if tier == 'Critical' else 'ðŸŸ  HIGH' if tier == 'High' else 'ðŸŸ¡ MEDIUM' if tier == 'Medium' else 'ðŸŸ¢ LOW'}")

print(f"\nðŸŸ¢ UPGRADE OPPORTUNITIES:")
for tier in ['Critical', 'High', 'Medium', 'Low']:
    count = (predictions['upgrade_potential'] == tier).sum()
    pct = 100 * count / n_constituents
    print(f"   {tier:12}: {count:4} constituents ({pct:5.1f}%)")

print(f"\nðŸ’Ž HIGH-VALUE AT-RISK: {high_risk_high_value} constituents need immediate retention focus")
print(f"   Total at-risk value: ${predictions[predictions['churn_risk'] > 0.6]['lifetime_value'].sum():,.0f}")
```

## 8. Actionable Recommendations Engine

```python
# Generate recommendations for each constituent
recommendations_data = []

for idx, row in predictions.head(50).iterrows():
    churn_risk = row['churn_risk']
    upgrade_prob = row['upgrade_propensity']
    ltv = row['lifetime_value']
    
    # Decision logic
    if churn_risk > 0.7 and ltv > 100:
        action = 'ðŸ”´ Save First'
        priority = 'CRITICAL'
    elif churn_risk > 0.6 and ltv > 50:
        action = 'ðŸŸ  Careful Ask'
        priority = 'HIGH'
    elif upgrade_prob > 0.7:
        action = 'ðŸŸ¢ Prime Target'
        priority = 'HIGH'
    elif upgrade_prob > 0.5:
        action = 'ðŸŸ¡ Consider Ask'
        priority = 'MEDIUM'
    else:
        action = 'ðŸ’™ Nurture'
        priority = 'LOW'
    
    recommendations_data.append({
        'ID': row['constituent_id'],
        'Churn Risk': f"{churn_risk:.1%}",
        'Upgrade Prob': f"{upgrade_prob:.1%}",
        'LTV': f"${ltv:,.0f}",
        'Recommended Action': action,
        'Priority': priority
    })

recommendations_df = pd.DataFrame(recommendations_data)

# Display summary
print("\n" + "="*100)
print("ACTIONABLE RECOMMENDATIONS SAMPLE")
print("="*100)
print(recommendations_df.to_string(index=False))

# Create action distribution
action_counts = recommendations_df['Recommended Action'].value_counts()

fig = go.Figure([
    go.Bar(
        y=action_counts.index,
        x=action_counts.values,
        orientation='h',
        marker=dict(
            color=['#F44336', '#FF9800', '#4CAF50', '#FFC107', '#2196F3'],
            line=dict(color='black', width=1)
        ),
        text=[f'{v} constituents' for v in action_counts.values],
        textposition='outside',
        hovertemplate='<b>%{y}</b><br>Count: %{x}<extra></extra>'
    )
])

fig.update_layout(
    title='ðŸ“‹ Recommendation Distribution: Campaign Strategy',
    xaxis_title='Number of Constituents',
    yaxis_title='Recommended Action',
    height=400,
    showlegend=False
)

fig.show()

print("\n" + "="*100)
print("INTERPRETATION:")
print("="*100)
print("ðŸ”´ SAVE FIRST    - High-value constituents at critical churn risk")
print("                  â†’ Action: Personalized retention call within 48 hours")
print("\nðŸŸ  CAREFUL ASK   - Medium-risk, upgrade candidates")
print("                  â†’ Action: Segment-specific retention + upgrade offer")
print("\nðŸŸ¢ PRIME TARGET  - High upgrade propensity")
print("                  â†’ Action: Premium upgrade path, special pricing")
print("\nðŸŸ¡ CONSIDER ASK  - Moderate upgrade interest")
print("                  â†’ Action: Email drip campaign, demonstrate value")
print("\nðŸ’™ NURTURE       - Low risk, low upgrade interest")
print("                  â†’ Action: Content-based engagement, maintain relationship")
```
